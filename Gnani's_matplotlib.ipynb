```python
#plottig a line - using "plot" method which is present in the "pyplot" module
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,6,5,7])
y=np.array([0,30,120,250])
plt.plot(x,y)
#to visualize the graph - "show method is used"
plt.show()
```

```python
#we can plot with symbols without line (where symbols should be encoded with sible quotes --> '')
x=np.array([0,5,10,40])
y=np.array([0,30,120,250])
plt.plot(x,y,'o')
plt.show()
plt.plot(x,y,'*')
plt.show()
```

```python
#default x=values [0,1,2,3,4,.......,n]
y=np.array([5,3,6,7])
plt.plot(y)
plt.show()
```

```python
#attribute
# marker- to highlight the points (x,y) , it takes values = "Any symbols.."
x=np.array([1,2,3,4])
y=np.array([5,3,6,7])
plt.plot(x,y,marker='o')
plt.show()
```

```python
#shortcut - fmt
#fmt - marker|line-tyoe|color , values are encoded with single quotes 
#marker size --> ms=<number>
#marker border color --> mec='<color>' 
x=np.array([1,2,3,4])
y=np.array([5,3,6,7])
plt.plot(x,y,'*:r',ms=20,mec='b')
plt.show()

#color reference = {'g'	- Green	, 'b'	- Blue	, 'c'	- Cyan	, 'm'	- Magenta	, 'y'	- Yellow	, 'k'	- Black	 , 'w'	- White}
```

```python
#marker inside color --> mfc='<color>'
x=np.array([1,2,3,4])
y=np.array([5,3,6,7])
plt.plot(x,y,ms=20,mec='b',mfc='green')
plt.show()
```

```python
#we can use hexa decimal values for coloring
#or directly names of the colors , ex: plt.plot(x,y,marker='*' ,ms=40, mec='hotpink', mfc = 'hotpink'/)
x=np.array([1,2,3,4])
y=np.array([5,3,6,7])
plt.plot(x,y,marker='*' ,ms=40,mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
#linestyle='<style name || style symbol >' , ex: plt.plot(x,y,linestyle='dashed')
#or ls='<style symbol || style name>'

#color='<color nmae || alphabet || hexa value>' , for line coloring

#linewidth=<number> - for line width
x=np.array([1,2,3,4])
y=np.array([5,3,6,7])
plt.plot(x,y,ls='--',color='r',linewidth=5)
plt.show()
```

```python
#multiple line in one graph
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
#assigning names to x-axis,y-axis and title to the graph
#methods - xlabel , ylabel ,title 
# fontdict - used to take dictionary of values
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("CSE", fontdict = font1)
plt.xlabel("X-AXIS", fontdict = font2)
plt.ylabel("Y-AXIS", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
#loc = '<position value>'  - specifies position value of the title
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("CSE", loc='left')
plt.plot(x, y)
plt.show()
```

```python
#grid() method - graph in the form of grids
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
plt.grid(color='g',ls='--',lw='1')
plt.show()
#axis='x' || axis='y' -- grids only for specified axis
plt.plot(x, y)
plt.grid(axis='x')
plt.show()
```

```python
#multiple graphs
#subplot[a,b,c] = a-row ,b-column, c-position.
#suptitle = used to set title for all graphs
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Graph-A")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Graph-B")

plt.suptitle("Graphs")
plt.show()
```

```python
#scatter plots

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

```python
#each color for a dot
x = np.array([5,7,8,7,2,17,2,9,4,11])
y = np.array([99,86,87,88,111,86,103,87,94,78])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown"])
plt.scatter(x, y, c=colors)

plt.show()
```

```python
#default colormap
#cmap() method
#colorbar() - mwthod to print cmap
x = np.array([5,7,8,7,2,17,2,9,4,11])
y = np.array([99,86,87,88,111,86,103,87,94,78])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
#we can assign array of sizes , by using attribute --> s=''
#aplha - transparency , range[0,1]
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

```python
#random -generates random values
#randint - generates random integer values
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 5*np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
#bar graphs
#color = name || alphabet || Hexa
#width = width of bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color='g' ,width = 0.1)
plt.show()
#for horizontal graphs -- barh()
plt.barh(x, y, color='g' ,height = 0.1)
```

```python
#histograms
#randint-(170,10,250)  , 250 values , around 170 , with deviation of 10
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

```python
#piechart

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
#explode =  how each wedge should far from center
myexplode = [0.2, 0, 0, 0]
#startangle - at which angle the chart should start
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True,startangle = 90)
plt.show()
```
